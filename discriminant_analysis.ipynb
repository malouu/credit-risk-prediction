Transformer nos variables catégorielles en variables continues.

```python
hot = pd.get_dummies(data[['credit_history', 'checking_status', 'purpose', 'savings_status', 
                        'employment', 'personal_status', 'other_parties', 
                        'installment_commitment', 'residence_since', 'property_magnitude', 
                        'other_payment_plans', 'housing', 'existing_credits', 
                        'job', 'num_dependents', 'own_telephone', 'foreign_worker']])
hot
 #La fonction pd.get_dummies() est une fonction de Pandas qui permet de transformer
 # des variables catégorielles en variables binaires (ou indicatrices) 
 # en utilisant la méthode OneHot encoding.

```

```python
data = pd.concat([data, hot], axis = 1)
data = data.drop(['credit_history', 'checking_status', 'purpose', 'savings_status', 
                        'employment', 'personal_status', 'other_parties', 
                        'installment_commitment', 'residence_since', 'property_magnitude', 
                        'other_payment_plans', 'housing', 'existing_credits', 
                        'job', 'num_dependents', 'own_telephone', 'foreign_worker'], axis = 1)
data
```

```python
# les colonne de la dataset après transformation
data.info()
```

```python
from sklearn import preprocessing
label_encoder_class = preprocessing.LabelEncoder()
data['class'] = label_encoder_class.fit_transform(data['class'])

# La fonction LabelEncoder attribue un numéro unique à chaque catégorie de la variable catégorielle, 
# ce qui permet à l'algorithme de traitement de données de comprendre comment les différentes catégories 
# sont liées les unes aux autres.
```

Séparation des variables caractéristiques et de la variable cible


```python
X = data.drop('class', axis = 1)
X = X.values
y = data['class']
```

```python
X
```

StandardScaler



```python
# Nous utiliserons StandardScaler pour mettre nos données à la même échelle
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_standard = scaler.fit_transform(X)
```

# Construction de modèle

Transformer les données en train et test (30% test)

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X_standard, y, test_size = 0.3, random_state = 0)
```

Équilibrage des classes

```python
# As we have many values ​​of a class and few values ​​of others, we will balance the classes 
# using OverSampling to make the classes balanced.
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=0)
X_train, y_train = ros.fit_resample(X_train, y_train)
```

```python
X_train
```

```python
X_train.shape, y_train.shape
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
clf = LinearDiscriminantAnalysis()
clf.fit(X_train, y_train)
```

Tester le modèle


```python
# Mesure de performance
clf.score(X_test,y_test)
```

```python
#matrice de confusion
pred = clf.predict(X_test)
from sklearn import metrics 
print(metrics.confusion_matrix(y_test,pred))
```
